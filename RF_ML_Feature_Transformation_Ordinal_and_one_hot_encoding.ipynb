# Ordinal Encoding
To ordinal encode, we need to



1.  Define a List of Ordinal Features
2.  Define a list of the ordered values for each feature
3.  Instantiate the Encoder
4.  Fit the Encoder on the training data
5.  Transform the training data
6.  Transform the test data




1. Define a List of Ordinal Features

```python
# define a list of columns to encode as ordinal
ordinal_cols = ['Exter Qual', 'Garage Qual', 'Exter Cond', 'Garage Cond']
ordinal_cols
```

2. Specify the order for each of the ordinal features

Note that, in this case, all of our ordinal features have the same classes. So we only have to define one list here to apply to all 4 of our features. If our ordinal feature had different values, we would need to define separate lists to match each feature.

```python
df.columns
print(df['Exter Qual'].unique())
print(df['Exter Cond'].unique())
print(df['Garage Qual'].unique())
print(df['Garage Cond'].unique())
```

```python
# Specifying the order of categories in quality/condition columns
ext_qual= ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
ext_cond= ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
garage_qual= ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
garage_cond = ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
#OR
#qual_cond_order - ['NA','Po', 'Fa', 'TA', 'Gd', 'Ex']
```

Now, we will create a list of our ordered lists that corresponds to each variable. (Since all four of our ordinal features have the same values, we will repeat that list 4 times).

```python
# Making the list of order lists for OrdinalEncoder
ordinal_category_orders=[ext_qual, ext_cond, garage_qual, garage_cond]
#OR
#ordinal_category_orders = [qual_cond_order, qual_cond_order, qual_cond_order, qual_cond_order]
```

3. Instantiate the Encoder

```python
# Instantiate the encoder and include the list of ordered values as an argument
ord_encoder = OrdinalEncoder(categories=ordinal_category_orders)
ord_encoder
```

4. Fit the Encoder on the Training Data

```python
# Fit the encoder on the training data
ord_encoder.fit(X_train_cat_imputed[ordinal_cols])
```

5. Transform the training data

```python
# Transform the training data
X_train_ordinal_enc = ord_encoder.transform(X_train_cat_imputed[ordinal_cols])
```

# One Hot encoding

OneHotEncoder follows the same process as our other transformers:

1. Define a list of nominal features to one-hot encode

2. Instantiate the encoder

3. Fit the OneHotEncoder ONLY on the categorical training data.

4. Transform the categorical training data to a one-hot encoded form.

5. Transform the categorical testing data to a one-hot encoded form.

1. Define a list of nominal Features

We only want to one-hot encode the categorical features that have not been encoded ordinally. We will make a new list of columns to encode using OneHotEncoder, which we will call ohe_cols.

We know that we have already encoded the ordinal features, so any remaining categorical feature will be encoded using OneHotEncoder.  We will define our list of features for one-hot encoding by dropping the ordinal columns.

```python
# saving list of categorical features to one-hot-encode
ohe_cols = X_train_cat_imputed.drop(columns=ordinal_cols).columns
ohe_cols

# X_train_cat_imputed[[['MS Zoning', 'Street', 'Alley', 'Utilities', 'Neighborhood',
#        'Bldg Type', 'House Style', 'Central Air', 'Gr Liv Area', 'Half Bath',
#        'Garage Type', 'Paved Drive', 'Fence', 'Date Sold']]]
```

2. Instantiate the Encoder

Scikit-Learn's OneHotEncoder class will do all of the one-hot encoding work for us, but we want to make some changes to the default settings when we instantiate the OneHotEncoder.

**sparse_output = False**

By default, it will return what is called a 'sparse matrix'. This is a form of data compression used when arrays are mostly filled with 0s. Instead of keeping track of so many 0s, the compressed version just has information about where the data is NOT 0. This is great for saving memory! But not great if we want to use the data in a new dataframe. We can get OneHotEncoder to return a normal array that is compatible with pandas, called a Dense array by specifying sparse=False inside our OneHotEncoder.

**handle_unknown ='ignore'**

By default, OneHotEncoder will throw an error if we try to transform data that has categories that were not present in the data it was fitted on and does not have columns for. We can pass the argument 'handle_unknown='ignore'' to tell it to ignore any categories that were not present during the fit step (values it encounters in the test set that were not present in the train set). In this case, all columns corresponding to that feature will have the value 0.

```python
# Instantiate one hot encoder
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
```

3. Fit the Encoder on the Training Data

```python
# Fit the OneHotEncoder on the training data
ohe_encoder.fit(X_train_cat_imputed[ohe_cols])
```

4. Transform the Training Data

```python
# Transform the training data
X_train_cat_ohe = ohe_encoder.transform(X_train_cat_imputed[ohe_cols])
X_train_cat_ohe
```

Notice that we now have many more columns!

You will see that "MS Zoning" has been replaced with 7 columns. Each column starts with the original feature name. Then, after an underscore, the value is given. If you look closely at the first row, you will see many 0.0, but there is a 1.0 present in the MS Zoning_RL column. This corresponds to this row having an original MS Zoning value of "RL."

5. Transform the testing data

```python
# Transform the testing data
X_test_cat_ohe = ohe_encoder.transform(X_test_cat_imputed[ohe_cols])
X_test_cat_ohe
```
