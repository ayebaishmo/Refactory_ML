What is one hot encoding?
One-hot encoding is a method used to convert non-ordered or norminal features in a numerical values a machine learninig can understand

It creates a new column for every category present in the feature
set the values of each columns to 1 if that row corresponds to the original category
set the value of each of the new columns to 0 if they donot

How to know whether to One hot encode or Ordinal encode?


If there is an order to the words, such as low Medium and high then we can encode them
as ordered numbers Ordinal encoding. Otherwise we can one hot encode them

```python
data = ['Red', 'Blue', 'Green', 'Blue']
```

```python
categories = list(set(data))
print(categories)
```

```python
One_hot_encode = []

for item in data:
    row = []
    for category in categories:
        if item == category:
            row.append(1)
        else:
            row.append(0)
    One_hot_encode.append(row)
print(One_hot_encode)
```

```python
One_hot_encode_col_columns = ['MS Zoning', 'Street', 'Alley', 'Utilities', 'Neighborhood', 'Bldg Type', 'House Style', 'Central Air', 'Garage Type', 'Paved Drive', 'Fence']
```

```python
from sklearn.preprocessing import OneHotEncoder
One_hot_encoder = OneHotEncoder()

One_hot_encoder.fit(X_train[One_hot_encode_col_columns])
```

```python
X_train_one_hot = One_hot_encoder.transform(X_train[One_hot_encode_col_columns])
```

```python
encoded_columns = One_hot_encoder.get_feature_names_out(One_hot_encode_col_columns)
encoded_columns
```

```python
X_train_one_hot_df = pd.DataFrame(
    X_train_one_hot.toarray(),
    columns=encoded_columns,
    index=X_train.index
)
```
